<h1 style='color: #0000b0'> SEABORN HEATMAP</h1>

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from funcs_armazem import get_bd_dados
import pandas as pd
pd.options.display.max_rows = 999
# plt.gcf().canvas.get_supported_filetypes() # PARA CONSULTAR FORMATOS SUPORTADOS POR pyplot.savefig()

bd_dados = get_bd_dados(5)
```

```python
semana_hora_142 = bd_dados['tqf'][
    ( bd_dados['tqf']['23_CIA'] == '142 CIA' )
    & ( bd_dados['tqf']['MES'] <= 5 )
].pivot_table(
    aggfunc=sum,
    index=['Faixa 1 Hora Fato'],
    columns='Dia da Semana Fato',
    values='Qtde Ocorrências'
).reindex(columns=[
    'SEGUNDA-FEIRA',
    'TERÇA-FEIRA',
    'QUARTA-FEIRA',
    'QUINTA-FEIRA',
    'SEXTA-FEIRA',
    'SÁBADO',
    'DOMINGO'
])
```

```python

fig = plt.figure(figsize=(12,9))
sns.heatmap(semana_hora_142, cmap='Reds', annot=True)
plt.yticks(rotation=0)
plt.xticks(rotation=45)
plt.tight_layout()
plt.title(
    'Furtos, por turnos/Dias da Semana, em 2020 (jan a mai) - 142ª Cia PM',
    fontsize=20
)
fig.align_ylabels()
plt.savefig('Mapa Calor Furtos - 142 Cia - Jan a Maio.pdf', bbox_inches='tight')
plt.show()
```
