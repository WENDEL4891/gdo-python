```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
cvs = pd.read_excel('files/Armazem/2020/00.Índice de Crimes Violentos_(TCV)_GDO_2020_TENTADO E CONSUMADO_extracao.xls', sheet_name='BD')
```

```python
cvs.columns
```

```python
sns.barplot(x='Dia da Semana Fato', y='Qtde Ocorrências', data=cvs, estimator=np.sum)
plt.xticks(rotation=70)
plt.tight_layout()
```

```python
sns.barplot(x='23_CIA', y='Qtde Ocorrências', data=cvs, estimator=np.sum)
plt.xticks(rotation=70)
plt.tight_layout()
```

```python
sns.barplot(x='Faixa 6 Horas Fato', y='Qtde Ocorrências', data=cvs, estimator=np.sum)
plt.xticks(rotation=70)
plt.tight_layout()
```

```python
sns.barplot(x='Município', y='Qtde Ocorrências', data=cvs[
    ~ cvs['Município'].isin(['DIVINOPOLIS', 'ITAUNA'])
], estimator=np.sum)
plt.xticks(rotation=70)
plt.tight_layout()
```

```python
cvs.columns
```

```python
cv_hora_semana = cvs.pivot_table(
    values='Qtde Ocorrências',
    index='Faixa 1 Hora Fato',
    columns='Dia da Semana Fato',
    aggfunc=sum,
    fill_value=0
).reindex([
    'SEGUNDA-FEIRA',
    'TERÇA-FEIRA',
    'QUARTA-FEIRA',
    'QUINTA-FEIRA',
    'SEXTA-FEIRA',
    'SÁBADO',
    'DOMINGO'
    
], axis=1)
```

```python
cv_hora_semana
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(cv_hora_semana, cmap='Reds', linecolor='black', linewidths=1, annot=True)

plt.xticks(rotation=70)
plt.tight_layout()
```

```python
cv_mes_semana = cvs[
    cvs['Mês Numérico Fato'] <= 4
].pivot_table(
    values='Qtde Ocorrências',
    index='Mês Numérico Fato',
    columns='Dia da Semana Fato',
    aggfunc=sum,
    fill_value=0
).reindex([
    'SEGUNDA-FEIRA',
    'TERÇA-FEIRA',
    'QUARTA-FEIRA',
    'QUINTA-FEIRA',
    'SEXTA-FEIRA',
    'SÁBADO',
    'DOMINGO'
    
], axis=1)
cv_mes_semana
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(cv_mes_semana, cmap='Reds', linecolor='white', linewidths=1, annot=True)
```

```python
cv_cia_semana = cvs.pivot_table(
    values='Qtde Ocorrências',
    index='23_CIA',
    columns='Dia da Semana Fato',
    aggfunc=sum,
    fill_value=0
).reindex([
    'SEGUNDA-FEIRA',
    'TERÇA-FEIRA',
    'QUARTA-FEIRA',
    'QUINTA-FEIRA',
    'SEXTA-FEIRA',
    'SÁBADO',
    'DOMINGO'
    
], axis=1)
cv_mes_semana
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(cv_cia_semana, cmap='Reds', linecolor='white', linewidths=1, annot=True)
```
