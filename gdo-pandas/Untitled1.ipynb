```python
import pandas as pd
```

```python
rats = pd.read_sql_table('tbl_rat', 'sqlite:///gdo.db')
```

```python
rats
```

```python
rats_itauna_2020 = rats[
    (rats['MUNICIPIO'] == 'ITAUNA')
    & (rats['ANO'] == 2020)
]
```

```python
ops_transito_itauna_2020 = pd.pivot_table(
    data=rats_itauna_2020\
    [
        ( rats_itauna_2020['NAT.DESCRICAO'].str.contains('TRANSITO') )
#         | ( rats_itauna_2020['NAT.DESCRICAO'].str.contains('OPERACAO') )
    ]
    ,
    values='RAT.NUM_ATIVIDADE',
    columns='MES',
    index='NAT.DESCRICAO',
    aggfunc='count'
)
```

```python
ops_transito_itauna_2020.loc[:,'TOTAL_POR NATUREZA'] = ops_transito_itauna_2020.sum(1)
```

```python
ops_transito_itauna_2020.loc['TOTAL POR MES'] = ops_transito_itauna_2020.sum()
```

```python
ops_transito_itauna_2020
```

```python
writer = pd.ExcelWriter('Operações Trânsito Itaúna 2020.xlsx', engine='xlsxwriter')
ops_transito_itauna_2020.to_excel(writer, sheet_name='Resumo por mês e natureza')
rats.to_excel(writer, sheet_name='RATS 2020 - 23 BPM')
writer.save()
```
